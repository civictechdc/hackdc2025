```python
import re

def get_email(comment_text):
    if not isinstance(comment_text, str):
        return None
    match = re.search(r'[\w.+-]+@[\w-]+\.[\w.-]+', comment_text)
    return match.group(0) if match else None
```

```python
emails = [email for text in df["comment"] if (email := get_email(text)) is not None] 
```

```python
extracted_domains = []
for email in emails:
    match = re.search(r"@([a-zA-Z0-9_.-]+)(?=\.com|\.edu|\.org|\.gov|\.net)", email)
    extracted_domains.append(match.group(1).lower() if match else None)
```

```python
df = df.with_columns(
pl.col("organization")\
  .str.to_lowercase()\
  .str.replace_all(r"\([^)]*\)", "")\
  .str.replace_all(r"[ .,]", "")\
  .str.replace_all(r"(?i)(llc|inc)", "")\
  .alias("organization")
)
```

```python
org_domains = df.get_column("organization").drop_nulls().to_list()
```

```python
from rapidfuzz import fuzz, process

def fuzzy_match(list1, list2):
    matches = []
    for item in list1:
        match, score, _ = process.extractOne(item, list2, scorer=fuzz.ratio)
        matches.append((item, match, score))

    matches.sort(key=lambda x: x[2], reverse=True)

    return matches

email_fuzzy_matches = fuzzy_match(org_domains, extracted_domains)
```

```python
from english_words import get_english_words_set
from textblob import TextBlob

# Regex for acronyms (e.g. NASA, FDA, etc.)
acronym_pattern = r'\b[A-Z]{2,}\b'

# Get lowercase English words set
english_words = get_english_words_set(['gcide'], lower=True)

# Optional acronyms to exclude
exclude_acronyms = {'BP', 'CEO', 'TV', 'PDF', 'ICU', 'JSON', 'MD', 'CT', 'MPH', 'MBA', 'RFI'}

# Define function to extract acronyms
def find_acronyms(text):
    if not isinstance(text, str):
        return []

    blob = TextBlob(text)
    tokens = blob.words

    acronyms = [
        token for token in tokens
        if re.match(acronym_pattern, token)
        and '-' not in token
        and '/' not in token
        and token.lower() not in english_words
        and token not in exclude_acronyms
    ]

    # Handle "LLC" patterns
    for i in range(len(tokens) - 1):
        if tokens[i + 1] == 'LLC':
            preceding_word = tokens[i]
            if preceding_word not in acronyms:
                acronyms.append(preceding_word)

    return acronyms

# Apply function to Polars DataFrame
df = df.with_columns(
    pl.col("comment").map_elements(find_acronyms, return_dtype=pl.List(pl.String)).alias("acronyms")
)
```
